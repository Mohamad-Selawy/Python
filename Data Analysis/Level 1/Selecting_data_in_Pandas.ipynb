## 6. Selecting Columns

Selecting columns is one of the most common operations. You can select a single column or multiple columns from a DataFrame.

### Selecting a Single Column

To select a single column, use square brackets `[]` with the column name as a string. This returns a Pandas Series.

````python
# Selecting the 'Salary' column
salary_column = df["Salary"]
print("Type of selected column:", type(salary_column))
print("\nFirst 5 rows of Salary column:")
display(salary_column.head())
````

### Selecting Multiple Columns

To select multiple columns, pass a **list of column names** inside the square brackets. This returns a new DataFrame.

````python
# Selecting 'Name' and 'Department' columns
name_department = df[["Name", "Department"]]
print("Type of selected columns:", type(name_department))
print("\nFirst 5 rows of Name and Department columns:")
display(name_department.head())
````

## 7. Selecting Rows Using Labels (`.loc`)

The `.loc` accessor is used for **label-based indexing**. This means you use the actual row and column labels to make selections. It is inclusive of both the start and end of the slice.

### Selecting a Single Row by Label

To select a single row, use `.loc[row_label]`.

````python
# Select the row with index label 3
row_3 = df.loc[3]
print("Row at index label 3:")
display(row_3)
````

### Selecting Multiple Rows by Label (Slice)

Use slicing with row labels: `df.loc[start_label:end_label]`. Both `start_label` and `end_label` are included.

````python
# Select rows from index label 2 to 6 (inclusive)
rows_2_to_6 = df.loc[2:6]
print("Rows from index label 2 to 6:")
display(rows_2_to_6)
````

### Selecting Specific Rows by Label (List)

Pass a list of specific row labels to select non-contiguous rows: `df.loc[[label1, label2, ...]]`.

````python
# Select rows with index labels 1, 4, and 8
specific_rows = df.loc[[1, 4, 8]]
print("Specific rows with index labels 1, 4, 8:")
display(specific_rows)
````

### Selecting Rows and Columns Together with `.loc`

You can specify both row and column labels using `df.loc[row_labels, column_labels]`.

````python
# Select rows from 0 to 5 and columns 'Name' and 'Salary'
subset_loc = df.loc[0:5, ["Name", "Salary"]]
print("Subset using .loc:")
display(subset_loc)
````

## 8. Selecting Rows Using Position (`.iloc`)

The `.iloc` accessor is used for **integer-location based indexing**. This means you use the integer position of rows and columns (0-based) to make selections. It is exclusive of the end of the slice (like standard Python list slicing).

### Selecting a Single Row by Position

To select a single row, use `df.iloc[row_position]`.

````python
# Select the row at integer position 3 (the 4th row)
row_at_pos_3 = df.iloc[3]
print("Row at integer position 3:")
display(row_at_pos_3)
````

### Selecting Multiple Rows by Position (Slice)

Use slicing with integer positions: `df.iloc[start_pos:end_pos]`. Note that `end_pos` is *exclusive*.

````python
# Select rows from integer position 2 up to (but not including) 6
rows_pos_2_to_5 = df.iloc[2:6]
print("Rows from integer position 2 to 5:")
display(rows_pos_2_to_5)
````

### Selecting Rows and Columns Together with `.iloc`

You can specify both row and column positions using `df.iloc[row_positions, column_positions]`.

````python
# Select rows from position 1 up to (but not including) 6,
# and columns from position 2 up to (but not including) 5
subset_iloc = df.iloc[1:6, 2:5]
print("Subset using .iloc:")
display(subset_iloc)
````

## 9. Difference Between `.loc` and `.iloc`

Understanding the distinction between `.loc` and `.iloc` is crucial for effective data selection in Pandas. Here's a comparison:

| Feature           | `.loc` (Label-based)                           | `.iloc` (Integer-location based)                 |
| :---------------- | :--------------------------------------------- | :----------------------------------------------- |
| **Indexing Type** | Uses actual row/column labels                  | Uses integer positions (0-based)                 |
| **Slicing End**   | **Inclusive** of the end label (`start:end` includes `end`) | **Exclusive** of the end position (`start:end` excludes `end`) |
| **Input Types**   | Labels (strings, numbers, datetimes, booleans) | Integers or lists/slices of integers             |
| **Use Case**      | When you know the specific labels of rows/columns you want | When you know the numerical positions of rows/columns you want |

Let's illustrate with examples:

````python
# Example: Select a single element

# Using .loc (label for row index 5, label for column 'Salary')
print(f"Value at df.loc[5, 'Salary']: {df.loc[5, 'Salary']}")

# Using .iloc (integer position for row 5, integer position for 'Salary' column - let's assume it's 5th column index 4)
# First, find the integer position of 'Salary' column
salary_col_pos = df.columns.get_loc('Salary')
print(f"Value at df.iloc[5, {salary_col_pos}]: {df.iloc[5, salary_col_pos]}")

print("\n---")

# Example: Slicing rows

# Using .loc (inclusive)
print("Rows df.loc[0:2] (includes index 2):")
display(df.loc[0:2])

# Using .iloc (exclusive)
print("Rows df.iloc[0:2] (excludes index 2, so only 0 and 1):")
display(df.iloc[0:2])

print("\n---")

# Example: Slicing rows and columns

# Using .loc (inclusive for both row and column labels)
print("df.loc[0:1, ['Name', 'Department']] (includes row 1, includes Department column):")
display(df.loc[0:1, ['Name', 'Department']])

# Using .iloc (exclusive for both row and column positions)
# 'Name' is col 1 (index 1), 'Department' is col 2 (index 2)
print("df.iloc[0:2, 1:3] (excludes row 2, excludes column at index 3):")
display(df.iloc[0:2, 1:3])
````

## 10. Fast Scalar Access (`.at` and `.iat`)

When you need to access a single value (scalar) in a DataFrame, `.at` and `.iat` provide faster access than `.loc` and `.iloc` respectively, especially for large DataFrames.

### `.at` (Label-based Scalar Access)

`.at` is used to get a single value by row label and column label.

````python
# Get the salary of the employee at index label 3
salary_at_3 = df.at[3, "Salary"]
print(f"Salary of employee at index 3 using .at: {salary_at_3}")
````

### `.iat` (Integer-location based Scalar Access)

`.iat` is used to get a single value by row integer position and column integer position.

````python
# Get the value at row integer position 3 and column integer position 4 (Age column)
# Let's verify column position first: df.columns.get_loc('Age')
age_col_pos = df.columns.get_loc('Age')
age_at_3_4 = df.iat[3, age_col_pos]
print(f"Age of employee at row position 3 and column position {age_col_pos} (Age) using .iat: {age_at_3_4}")
````

### When to use `.at` and `.iat`?

Use `.at` or `.iat` when you need to access or modify a **single cell** and performance is critical. For general selection of slices or multiple values, `.loc` and `.iloc` are more appropriate.

## 11. Boolean Indexing

Boolean indexing (also known as Boolean filtering) is a powerful way to select rows based on whether they meet specific conditions. You provide a Series of `True`/`False` values to the DataFrame, and only rows corresponding to `True` will be selected.

### Example: Employees older than 30

````python
# Create a boolean Series
condition_age = df["Age"] > 30
print("Boolean Series for Age > 30:")
display(condition_age.head())

# Use the boolean Series to filter the DataFrame
employees_older_than_30 = df[condition_age]
print("\nEmployees older than 30:")
display(employees_older_than_30.head())
````

### Example: Salary greater than 70000

````python
high_salary_employees = df[df["Salary"] > 70000]
print("Employees with Salary > 70000:")
display(high_salary_employees.head())
````

### Example: Department equals 'IT'

````python
it_department_employees = df[df["Department"] == "IT"]
print("Employees in the IT Department:")
display(it_department_employees.head())
````

### Example: City equals 'New York'

````python
ny_employees = df[df["City"] == "New York"]
print("Employees living in New York:")
display(ny_employees.head())
````

### Example: Experience greater than 5 years

````python
experienced_employees = df[df["Experience"] > 5]
print("Employees with more than 5 years of experience:")
display(experienced_employees.head())
````

## 12. Multiple Conditions

You can combine multiple boolean conditions using logical operators. **Crucially, each condition must be enclosed in parentheses.**

*   **AND (`&`)**: Both conditions must be `True`.
*   **OR (`|`)**: At least one condition must be `True`.
*   **NOT (`~`)**: Negates a condition (flips `True` to `False` and vice versa).

### Example: Employees older than 30 AND salary greater than 70000

````python
senior_high_earners = df[(df["Age"] > 30) & (df["Salary"] > 70000)]
print("Employees older than 30 AND salary > 70000:")
display(senior_high_earners.head())
````

### Example: Employees in IT OR HR department

````python
it_or_hr_employees = df[(df["Department"] == "IT") | (df["Department"] == "HR")]
print("Employees in IT OR HR department:")
display(it_or_hr_employees.head())
````

### Example: Employees NOT in the IT department

````python
not_it_employees = df[~(df["Department"] == "IT")]
print("Employees NOT in the IT department:")
display(not_it_employees.head())
````

### Importance of Parentheses

Parentheses are essential because of Python's operator precedence. Without them, the `&` and `|` operators would be evaluated before the comparison operators (`>`, `==`, etc.), leading to errors or incorrect results.

## 13. Using `isin()`

The `.isin()` method is a convenient way to filter a DataFrame when you want to select rows where a column's value is one of several possibilities. It's much cleaner than using multiple `OR` conditions.

````python
# Select employees whose department is either 'IT' or 'HR'
employees_in_it_or_hr_isin = df[df["Department"].isin(["IT", "HR"])]
print("Employees in IT or HR using .isin():")
display(employees_in_it_or_hr_isin.head())
````

### Practical Applications

*   **Filtering by categories**: Easily select data belonging to a predefined set of categories (e.g., specific product types, regions).
*   **Excluding values**: Combine with `~` (NOT) to exclude certain values (e.g., `df[~df['Status'].isin(['Retired', 'Terminated'])]`).

## 14. Using `between()`

The `.between()` method is used to select rows where a numeric column's value falls within a specified range (inclusive).

````python
# Select employees with salary between 50000 and 80000 (inclusive)
salary_range_employees = df[df["Salary"].between(50000, 80000)]
print("Employees with Salary between 50000 and 80000:")
display(salary_range_employees.head())
````

## 15. String Filtering

Pandas provides a `.str` accessor for Series that contain string data, allowing you to perform string-specific operations for filtering.

### `str.startswith()`

Selects rows where the string in a column starts with a specified prefix.

````python
# Select employees whose names start with 'A'
names_start_a = df[df["Name"].str.startswith("A")]
print("Employees whose names start with 'A':")
display(names_start_a)
````

### `str.endswith()`

Selects rows where the string in a column ends with a specified suffix.

````python
# Select employees whose names end with 'n'
names_end_n = df[df["Name"].str.endswith("n")]
print("Employees whose names end with 'n':")
display(names_end_n)
````

### `str.contains()`

Selects rows where the string in a column contains a specified substring. You can also use regular expressions.

````python
# Select employees whose city name contains 'New'
city_contains_new = df[df["City"].str.contains("New", case=False)] # case=False for case-insensitive search
print("Employees whose city name contains 'New':")
display(city_contains_new)
````

## 16. Common Mistakes

Even experienced users can sometimes trip up on Pandas indexing. Being aware of these common pitfalls can save you time and frustration.

### Mistake 1: Using parentheses instead of square brackets for column selection

Pandas uses square brackets `[]` for column selection, not parentheses `()`.

````python
# Incorrect way
# try:
#     df("Name")
# except TypeError as e:
#     print(f"Error: {e}")

# Correct way
correct_name_selection = df["Name"]
print("Correct column selection:")
display(correct_name_selection.head())
````

### Mistake 2: Forgetting parentheses around Boolean conditions

When combining multiple conditions with `&` or `|`, each individual condition must be wrapped in parentheses.

````python
# Incorrect way (will raise ValueError due to operator precedence)
# try:
#     df[df["Age"] > 30 & df["Department"] == "IT"]
# except ValueError as e:
#     print(f"Error: {e}")

# Correct way
correct_multi_condition = df[(df["Age"] > 30) & (df["Department"] == "IT")]
print("Correct multiple conditions selection:")
display(correct_multi_condition.head())
````

### Mistake 3: Confusing `.loc` and `.iloc` (especially with slicing)

Remember `.loc` is label-inclusive, `.iloc` is position-exclusive for the end of the slice.

````python
# If you want rows with index labels 0, 1, 2:
print("Using .loc[0:2] (inclusive of label 2):")
display(df.loc[0:2])

# If you want rows at integer positions 0, 1, 2:
print("Using .iloc[0:3] (exclusive of position 3):")
display(df.iloc[0:3])
````

### Mistake 4: Selecting multiple columns incorrectly

When selecting multiple columns, you must pass a *list* of column names inside the square brackets, not just comma-separated strings.

````python
# Incorrect way (will raise a KeyError)
# try:
#     df["Name", "Department"]
# except KeyError as e:
#     print(f"Error: {e} - Expected a list of column names for multiple selection")

# Correct way
correct_multi_column = df[["Name", "Department"]]
print("Correct multiple column selection:")
display(correct_multi_column.head())
````
