## What is Probability?

Probability is the measure of the likelihood that an event will occur. It ranges between **0** (an impossible event) and **1** (a certain event).

### Basic Terms:
- **Experiment**: An action that leads to one or more outcomes (e.g., flipping a coin).
- **Sample Space (S)**: Set of all possible outcomes.
- **Event (E)**: A subset of the sample space.

### Formula:
```python
P(E) = Number of favorable outcomes / Total number of possible outcomes
```

### Example:
```python
# Probability of getting a 'Head' in a coin toss
P_Head = 1 / 2  # 0.5
```


## Types of Distributions

### 1. **Uniform Distribution**
All outcomes are equally likely.
- Example: Rolling a fair die (1–6)
- Probability of any value = 1/6

````python
# @title
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the uniform distribution
low = 0   # Lower bound
high = 10 # Upper bound
size = 100000 # Number of samples

# Generate random samples from a uniform distribution
samples = np.random.uniform(low, high, size)

# Calculate the theoretical PDF value
theoretical_pdf_value = 1 / (high - low)

# Plotting the histogram to visualize the distribution
plt.figure(figsize=(8, 6))
plt.hist(samples, bins=20, density=True, color='skyblue', edgecolor='black', alpha=0.7, label='Sampled Data')

# Plot the theoretical PDF as a straight line
plt.plot([low, high], [theoretical_pdf_value, theoretical_pdf_value], 'r--', linewidth=2, label='Theoretical Distribution')

plt.title(f'Uniform Distribution (low={low}, high={high})')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(axis='y', alpha=0.75)
plt.legend()
plt.show()
````

### 2. **Binomial Distribution**
Used when there are exactly two outcomes (success/failure) across **n** trials.
- Parameters: `n` (number of trials), `p` (probability of success)

````python
# @title
from scipy.stats import binom

# Parameters for the Binomial Distribution
n = 10  # Number of trials
p = 0.5 # Probability of success in each trial

# Generate random samples
samples = np.random.binomial(n, p, size=10000)

# Calculate PMF for theoretical comparison
k = np.arange(0, n + 1) # Possible number of successes
pmf = binom.pmf(k, n, p)

plt.figure(figsize=(8, 6))
plt.hist(samples, bins=np.arange(-0.5, n + 1.5, 1), density=True, color='lightcoral', edgecolor='black', alpha=0.7, label='Sampled Data Histogram')

plt.title(f'Binomial Distribution (n={n}, p={p})')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.xticks(k)
plt.grid(axis='y', alpha=0.75)
plt.legend()
plt.show()
````

### 3. **Normal Distribution (Gaussian)**
A symmetric, bell-shaped distribution defined by its **mean** (µ) and **standard deviation** (σ).
- Many real-world datasets follow this.



````python
# @title
from scipy.stats import norm

# Parameters for the Normal Distribution
mean = 0    # Mean (µ)
std_dev = 1 # Standard deviation (σ)

# Generate random samples
samples = np.random.normal(mean, std_dev, size=100000)

# Create a range for the theoretical PDF
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 100)
# Calculate the theoretical PDF
pdf = norm.pdf(x, mean, std_dev)

plt.figure(figsize=(8, 6))
plt.hist(samples, bins=50, density=True, color='lightgreen', edgecolor='black', alpha=0.7, label='Sampled Data Histogram')
plt.plot(x, pdf, 'r-', linewidth=2, label='Theoretical PDF')

plt.title(f'Normal Distribution (Mean={mean}, Std Dev={std_dev})')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.grid(axis='y', alpha=0.75)
plt.legend()
plt.show()
````
